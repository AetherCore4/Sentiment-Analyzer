```python
import numpy as np 
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('train.txt', sep=';', header=None, names=['text', 'emotion'])
```

```python
df
```

```python
df.isnull().sum()
```

```python
unique_emotions = df['emotion'].unique()
emotion_number = {}
number_emotion = {}
i = 0
for emo in unique_emotions:
    emotion_number[emo] = i
    number_emotion[i] = emo
    i += 1

df['emotion'] = df['emotion'].map(emotion_number)
```

```python
df
```

```python
df['text'] = df['text'].apply(lambda x:x.lower())
```

```python
df
```

```python
import string
```

```python
def remove_punc(txt):
    return txt.translate(str.maketrans('', '', string.punctuation))
```

```python
df['text'] = df['text'].apply(remove_punc)
```

```python
df
```

```python
def remove_num(txt):
    new = ""
    for i in txt:
        if not i.isdigit():
            new += i
    return new

df['text'] = df['text'].apply(remove_num)
```

```python
def remove_emoji(txt):
    new= ""
    for i in txt:
        if i.isascii():
            new+=i
    return new
df['text'] = df['text'].apply(remove_emoji)
```

```python
from nltk.corpus import stopwords
```

```python
nltk.download('punkt')
nltk.download('stopwords')
```

```python
stop_words = set(stopwords.words('english'))
```

```python
stop_words
```

```python
len(stop_words)
```

```python
df.loc[1]['text']
```

```python
def remove_stopwords(txt):
    words = txt.split()
    cleaned_txt = []
    for i in words:
        if not i in stop_words:
            cleaned_txt.append(i)
    return ' '.join(cleaned_txt)
```

```python
df['text'] = df['text'].apply(remove_stopwords)
```

```python
df.loc[1]['text']
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(df['text'],df['emotion'],test_size=0.20,random_state=42)
```

```python
len(X_train)
```

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer

from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score

bow_vectorizer = CountVectorizer()
X_train_bow = bow_vectorizer.fit_transform(X_train)
X_test_bow = bow_vectorizer.transform(X_test)


nb_model = MultinomialNB()
nb_model.fit(X_train_bow, y_train)


pred_bow = nb_model.predict(X_test_bow)
print(accuracy_score(y_test, pred_bow))
```

```python
pred_bow
```

```python
tfidf_vectorizer = TfidfVectorizer()
X_train_tfidf = tfidf_vectorizer.fit_transform(X_train)
X_test_tfidf = tfidf_vectorizer.transform(X_test)
```

```python
nb2_model = MultinomialNB()
nb2_model.fit(X_train_tfidf,y_train)
```

```python
y_pred = nb2_model.predict(X_test_tfidf)
```

```python
print(accuracy_score(y_test,y_pred))
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logistic_model = LogisticRegression(max_iter=1000)
```

```python
logistic_model.fit(X_train_tfidf,y_train)
```

```python
log_pred = logistic_model.predict(X_test_tfidf)
```

```python
print(accuracy_score(log_pred,y_test))
```

```python
# Save trained models for server.py
import pickle

# Save only picklable objects
models_data = {
    'logistic_model': logistic_model,
    'nb2_model': nb2_model,
    'tfidf_vectorizer': tfidf_vectorizer,
    'emotion_number': emotion_number,
    'number_emotion': number_emotion,
    'df': df
}

with open('models_data.pkl', 'wb') as f:
    pickle.dump(models_data, f)

print("✓ Models saved successfully!")
print("Now run: python server.py")
print("Now run: python server.py")
```
